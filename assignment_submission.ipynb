```python
df = pd.read_csv('C:/Users/Carson Colyer/Downloads/sharks.csv', low_memory=False)
df.head()
# df.columns.tolist()
```

```python
df['Type'].value_counts()
```

```python
type_col = df['Type']

type_col = type_col.replace(['Sea Disaster', 'Boat', 'Boating', 'Boatomg'], 'Watercraft') # Replace Boat, Boating, Sea Disaster, etc., with 'Watercraft'

# Group other infrequent or invalid values into a single category
infrequent_categories = ['Invalid', 'Questionable', 'Unconfirmed', 'Unverified', 'Under investigation']
type_col = type_col.replace(infrequent_categories, np.nan)

df['Type'] = type_col # Reassign the cleaned column

df = df.dropna(subset=['Type'])

print(df['Type'].value_counts()) # Display the cleaned value counts
```

```python

df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # All unclean values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') # All unclean values
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
```

```python
df = pd.read_csv("C:/Users/Carson Colyer/Downloads/VirginiaPretrialData2017.csv", low_memory=False)
df.head()
# df.columns.tolist()
```

```python
release = df['WhetherDefendantWasReleasedPretrial']
# print(release.unique(),'\n')
# print(release.value_counts(),'\n')
release = release.replace(9, np.nan) # Treat 9 as missing

release = release.map({1: 1, 0: 0, np.nan: np.nan}) # Convert to dummy (binary) variable: Released = 1, Not released = 0 (Used ChatGPT here)

df['WhetherDefendantWasReleasedPretrial'] = release

df = df.dropna(subset=['WhetherDefendantWasReleasedPretrial'])


print(df['WhetherDefendantWasReleasedPretrial'].value_counts(dropna=False))
```

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

length = pd.to_numeric(length,errors='coerce') # Coerce length to numeric
length_NA = length.isnull() # Create a missing dummy
print(np.sum(length_NA),'\n') # 9k missing values out of 23k

print(pd.crosstab(length_NA, type), '\n') # Category 4 is cases where the charges were dismissed

length = length.mask( type == 4, 0) # Replace sentence length with 0 for cases where charges were dismissed (type 4)

length = length.mask( type == 9, np.nan) # Replace sentence length with NaN for ambiguous or unknown case types (type 9)

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')

df['ImposedSentenceAllChargeInContactEvent'] = length
```
